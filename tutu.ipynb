```python
# Чтение geojson с регионами России
geo_data = 'https://raw.githubusercontent.com/codeforgermany/click_that_hood/main/public/data/russia.geojson'

for i in list(df_main.columns)[1:]:
    m = folium.Map(location=[61.5240, 105.3188], zoom_start=3)
    folium.Choropleth(
        geo_data=geo_data,
        data=df_main,
        columns=['region', i],
        key_on='feature.properties.name',
        fill_color='OrRd',
        fill_opacity=0.5,
        line_opacity=0.2,
        legend_name='Значения по регионам'
    ).add_to(m)

    m.save(f"res_defult_data/res_{i}.html")
```
